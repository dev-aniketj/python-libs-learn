```python
import pandas as pd

df = pd.read_csv('cities.csv')
df
```

```python
# acending order sorting
df.sort_values(by='LatD')
```

```python
# descending order sorting
df.sort_values(by='LatD', ascending=False)
```

```python
# create new file and export and commit changes in it.
df = df.sort_values(by='LatD', ascending=False)
df.to_csv("myFile.csv", index=False)
```

```python

df2 = pd.read_csv('myFile.csv')
df2
```

```python
# join two files

# column wise
df_join = pd.concat([df, df2], axis=1)
df_join
```

```python
# row wise
df_join = pd.concat([df, df2], axis=0)
df_join
```

```python
# groupby()

df.groupby(df['LatD']).mean()
```
